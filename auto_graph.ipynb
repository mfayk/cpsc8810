```python
import pandas 
import seaborn
import csv
import matplotlib.pyplot as plt

def make_csv(intxt):
    read_file = pandas.read_csv(intxt)
    read_file.to_csv(intxt[:-3] + 'csv', index=None)


def print_graphs(intxt):
    df = pandas.read_csv(intxt[:-3] + 'csv')
    plt = seaborn.barplot(x='method', y='time [ms]', hue="block Size", ci=False, data=df)
     plt.set_yscale('log')
```

```python
make_csv('metrics.txt')
print_graphs('metrics.txt')
```
